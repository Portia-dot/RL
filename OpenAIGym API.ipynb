```python
import random
from typing import List
```

```python
class Environment:
    def __init__(self):
        #This limit the number of steps the agent can take to interact with the environment
        self.steps_left = 10
    def get_observation(self) -> List[float]:  #-> indicates the function is expected to return a list of floats
        #This function returns the observation of the environment
        return [0.0, 0.0, 0.0, 0.0]
    def get_actions(self) -> List[int]:
        #There are only two actions possible in this environment, so we return a list of two integers
        return [0, 1]
    def is_done(self) -> bool:
        return self.steps_left == 0

    def action(self, action: int) -> float:
        if self.is_done():
            raise Exception("Tried to take an action in a done environment!")
        self.steps_left -=  1
        return random.random()
```

```python
class Agent:
    def __init__(self):
        self.total_rewards = 0.0

    def step(self, env: Environment):
        current_observation = env.get_observation()
        print(f'Current Observation Is {current_observation}')
        actions = env.get_actions()
        print(f'Available Actions Are {actions}')
        rewards = env.action(random.choice(actions))
        self.total_rewards += rewards
```

```python
env = Environment()
agent = Agent()
while not env.is_done():
    agent.step(env)
print(f'Total Reward Got Is {agent.total_rewards}')
```
